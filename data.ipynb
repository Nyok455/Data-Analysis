```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
# Sample data
x = np.linspace(0, 10, 100)  # Generate 100 points from 0 to 10
y1 = np.sin(x)                # Sine wave
y2 = np.cos(x)                # Cosine wave
# Create a figure and axis
plt.figure(figsize=(12, 6))
# Plotting the sine wave
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
# Plotting the cosine wave as a scatter plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.scatter(x, y2, label='Cosine Wave', color='red', marker='o')
plt.title('Cosine Wave (Scatter Plot)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
# Show the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

Data Processing

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Sample data creation (you can replace this with loading your dataset)
data = {
    'Product': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Sales': [250, 150, 300, np.nan, 400, 200, 350, 500],
    'Profit': [50, 20, 60, 30, 80, np.nan, 70, 100],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West']
}
# Creating a DataFrame
df = pd.DataFrame(data)
# Display the original DataFrame
print("Original DataFrame:")
print(df)
# Data Processing
# 1. Handling missing values by filling with the mean
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
df['Profit'] = df['Profit'].fillna(df['Profit'].mean())
# 2. Adding a new column for Profit Margin
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100
# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)
# 3. Summarizing data
summary = df.describe()
print("\nSummary Statistics:")
print(summary)
# 4. Grouping by Region and calculating total sales and profit
grouped = df.groupby('Region').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
print("\nGrouped Data by Region:")
print(grouped)
# Visualization
plt.figure(figsize=(12, 6))
# Bar Plot for Sales by Region
plt.subplot(1, 2, 1)
plt.bar(grouped['Region'], grouped['Sales'], color='blue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
# Bar Plot for Profit by Region
plt.subplot(1, 2, 2)
plt.bar(grouped['Region'], grouped['Profit'], color='green')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.tight_layout()
plt.show()
```
